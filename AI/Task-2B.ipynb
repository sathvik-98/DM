```python
import ttg
B = [0.001,0.999]
E = [0.002,0.998]
A = [[[0.95,0.94],[0.29,0.001]], [[0.05,0.06],[0.71,0.999]]]
J = [[0.9,0.05],[0.1,0.95]]
M = [[0.7,0.01],[0.3,0.99]]

def computeProbability(b, e, a, j, m):
    if b == True:
        p_b = B[0]
    else:
        p_b = B[1]
    if e == True:
        e_b = E[0]
    else:
        e_b = E[1]
    if a == True:
        a_b = A[0][B.index(p_b)][E.index(e_b)]
        k = 0
    else:
        a_b = A[1][B.index(p_b)][E.index(e_b)]
        k = 1
    if j == True:
        j_b = J[0][k]
    else:
        j_b = J[1][k]
    if m == True:
        m_b = M[0][k]
    else:
        m_b = M[1][k]

    return (p_b*e_b*a_b*j_b*m_b)

def values_list(ip):
    b = 'U';e = 'U';a = 'U';j = 'U';m = 'U'
    for i in ip:
        if i == 'Bt':
            b = True
        elif i == 'Bf':
            b = False
        elif i == 'Et':
            e = True
        elif i == 'Ef':
            e = False
        elif i == 'At':
            a = True
        elif i == 'Af':
            a = False
        elif i == 'Jt':
            j = True
        elif i == 'Jf':
            j = False
        elif i == 'Mt':
            m = True
        elif i == 'Mf':
            m = False

    return [b,e,a,j,m]

def truthTable(ip):
    dic = {0:'b',1:'e',2:'a',3:'j',4:'m'}
    tt = ttg.Truths([dic[i] for i in ip ])
    return tt


def bnet():  
    sum = 0;sum1=0;sum2=0
    ip = list(input().split(' '))
    len1  = len(ip)
    for i in ip:
        if i == 'given':
            l1 = [i for i in ip]
            l2 = [i for i in ip[ip.index('given')+1:len1]]
            nume_list = values_list(l1)
            denom_list = values_list(l2)
            break
        else:
            nume_list = values_list(ip)
            denom_list = 'no denominator'
    
    if denom_list == 'no denominator':
        indices1 = [i for i, x in enumerate(nume_list) if x == "U"]
        tt = truthTable(indices1)
        list1 = tt.base_conditions
        for j in range(0,len(list1)):
            for i in range(0,len(indices1)):
                nume_list[indices1[i]] = tt.base_conditions[j][i]
            sum = sum + computeProbability(nume_list[0],nume_list[1],nume_list[2],nume_list[3],nume_list[4])
    else:
        indices1 = [i for i, x in enumerate(nume_list) if x == "U"]
        indices2 = [i for i, x in enumerate(denom_list) if x == "U"]
        tt1 = truthTable(indices1)
        list1 = tt1.base_conditions
        for j in range(0,len(list1)):
            for i in range(0,len(indices1)):
                nume_list[indices1[i]] = tt1.base_conditions[j][i]
            sum1 = sum1 + computeProbability(nume_list[0],nume_list[1],nume_list[2],nume_list[3],nume_list[4])
        tt2 = truthTable(indices2)
        list2 = tt2.base_conditions
        for j in range(0,len(list2)):
            for i in range(0,len(indices2)):
                denom_list[indices2[i]] = tt2.base_conditions[j][i]
            sum2 = sum2 + computeProbability(denom_list[0],denom_list[1],denom_list[2],denom_list[3],denom_list[4])
        sum = sum1/sum2
    return sum

bnet()
```
